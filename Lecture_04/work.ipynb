```julia
using DelimitedFiles
using PyPlot
```

```julia
url = "https://www.phys.uconn.edu/~rozman/Courses/P2200_25F/downloads/trinity-test.data"
```

```julia
obsevations = download(url)
```

```julia
dat = readdlm(obsevations, comments=true, comment_char='#')
```

```julia
t = dat[:,1]
```

```julia
radius = dat[:,2]
```

```julia
plot(t, radius);
```

```julia
loglog(t, radius, "bo")
```

```julia
loglog(t, radius, "ro", label="measurements")
grid(true)
xlabel("time (seconds)")
ylabel("radius (meters)")
title("Trinity Test Data (Observational)")
legend()
```

```julia
const beta = 2/5
rad(t,alpha) = alpha * t^beta ;
nexcl = 2
np = length(t)
ind = (nexcl+1):np
```

```julia
tlogbar = sum(log.(t[ind])) / length(ind)
rlogbar = sum(log.(radius[ind])) / length(ind)
alphalog = rlogbar - beta * tlogbar
alpha = exp(alphalog)
println("Î± = $(round(alpha, sigdigits=4))")
```

```julia
loglog(t, radius, "ro", label="Measurements")
loglog(t, rad.(t, alpha), "g", "-", label="Line of Best Fit")

grid(true)
xlabel("time (ms)")
ylabel("radius as a function of time (m)")
title("Trinity fireball radius as a function of time")
legend();
```
