```julia

"""
    ans = mytrapezoids(fun, a, b, n)

Numerically evaluate the integral int_a^b fun(x) dx using the
trapezoidal rule: I = h/2*(f_1 + 2f_2 + ... + 2f_{n-1} + f_n),
where h = (b - a)/(n - 1), x_i = a + (i - 1)*h, f_i = fun(x_i).
"""
function mytrapezoids(fun, a, b, n)
    h = (b-a)/(n-1)
    s1 = fun(a) + fun(b)
    s2 = 0.0
    for i in 2:(n-1)
        s2 += fun(a + (i-1)*h)
    end
    return h/2*(s1+2*s2)
end
```

```julia
fun1(x) = 1/x
a1 = 1
b1 = 2
exact1 = log(2)
np1 = 20
```

```julia
result1 = mytrapezoids(fun1, a1, b1, np1)
difference = abs(result1 - exact1)
print("Exact: ",round(exact1; sigdigits=2),"\nFunction test: ", round(result1; sigdigits=2),"\nDifference: ", round(difference; sigdigits=2))
```

```julia
fun2(x) = exp(x)
a2 = 0.0
b2 = 1.0
exact2 = exp(1)-1.0
np2 = 100
```

```julia
result2 = mytrapezoids(fun2, a2, b2, np2)
difference = abs(result2 - exact2)
print("Exact: ", round(exact2; sigdigits=2),"\nFunction test: ", round(result2; sigdigits=2),"\nDifference: ", round(difference, sigdigits=2))
```

```julia
fun3(x) = sin(x)
a3 = 0.0
b3 = pi
exact3 = 2
np3 = 20
```

```julia

n_exp = 10
hh = zeros(n_exp);               #creating an array with the same number of elements as the value of n_exp
abs_errors = zeros(n_exp);       #same thing as hh
```

```julia
for i = 1:n_exp
    np = 2^i+1
    hh[i] = (b3-a3)/(np-1)
    abs_errors[i] = abs(mytrapezoids(fun3, a3, b3, np) - exact3)
end
```

```julia
abs_errors
```

```julia

loglog(hh, abs_errors, marker=".", label="Numerical Experiment")
loglog(hh, hh.^2, linestyle="dashed", label=L"h^2")
grid(true)
xlabel("h")
ylabel("Absolute Error")
title("Error of the Trapezoidal method of Integral Approximation")
legend();
```
